## 이상치 제거 (Outlier)

```python
data = data[data["Age"] <= 100]
```

```python
def age_func(x):
    if x>80:
        return 80
    else:
        return x
```

```python
data["Age"] = data["Age"].apply(age_func)
```

```python
def new_age_func(x):
    if x > 70:
        return 70
    else:
        return x
```

```python
data["Age"].apply(new_age_func)
```

```python
data["Age"].apply(lambda x: 70 if x > 70 else x)
```
