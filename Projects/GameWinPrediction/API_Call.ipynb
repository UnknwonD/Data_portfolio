```python
import requests
from urllib import parse
import pprint
pp = pprint.PrettyPrinter(indent=4)
api_key = "mykey" # 하루마다 갱신됨.

request_header = {
                    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.164 Safari/537.36",
                    "Accept-Language": "ko,en-US;q=0.9,en;q=0.8,es;q=0.7",
                    "Accept-Charset": "application/x-www-form-urlencoded; charset=UTF-8",
                    "Origin": "https://developer.riotgames.com",
                    "X-Riot-Token": api_key
                }
# 소환사 정보 불러오는 함수
def summoner_v4_by_summoner_name(summonerName):
    encodingSummonerName = parse.quote(summonerName) # 한국어 닉네임 전처리
    url = f"https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/{encodingSummonerName}"
    return requests.get(url, headers=request_header).json()

def summoner_v4_by_accountId(accountId):
    url = f"https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-account/{accountId}"
    return requests.get(url, headers=request_header).json()

def summoner_v4_by_puuid(puuid):
    url = f"https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-puuid/{puuid}"
    return requests.get(url, headers=request_header).json()


def summoner_v4_summoner_name(id):
    url = f"https://kr.api.riotgames.com/lol/summoner/v4/summoners/{id}"
    return requests.get(url, headers=request_header).json()

# 랭크 불러오는 함수
def league_v4_id(encryptedSummonerId):
    url = f"https://kr.api.riotgames.com/lol/league/v4/entries/by-summoner/{encryptedSummonerId}"
    return requests.get(url, headers=request_header).json()

# 챔피언 숙련도 불러오는 함수
def mastery_by_id(encryptedSummonerId, line=""):
    url = f"https://kr.api.riotgames.com/lol/champion-mastery/v4/scores/by-summoner/{encryptedSummonerId}"
    return requests.get(url, headers=request_header).json()

# 현재 진행중인 게임 정보
def spectator(encryptedSummonerId):
    url = f"https://kr.api.riotgames.com/lol/spectator/v4/active-games/by-summoner/{encryptedSummonerId}"
    return requests.get(url, headers=request_header).json()

# match id 불러오기 위해 사용하는 함수
def match_v5_get_list_match_id(puuid, start, count):
    url = f"https://asia.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start={start}&count={count}"
    return requests.get(url, headers=request_header).json()



# matchid로 결과 확인하는 함수
def match_v5_get_match_history(matchId):
    url = f"https://asia.api.riotgames.com/lol/match/v5/matches/{matchId}"
    return requests.get(url, headers=request_header).json()

# 매치 타임라인보는 함수
def match_v5_get_match_timeline(matchId):
    url = f"https://asia.api.riotgames.com/lol/match/v5/matches/{matchId}/timeline"
    return requests.get(url, headers=request_header).json()

def summonerIdSearch(summonerId):
    url = f"https://asia.api.riotgames.com/lol/clash/v1/players/by-summoner/{summonerId}"
    return requests.get(url, headers=request_header).json()

# 챔피언 숙련도
def mastery(encryptedSummonerId):
    url = f"https://asia.api.riotgames.com/lol/champion-mastery/v4/champion-masteries/by-summoner/{encryptedSummonerId}"
    return requests.get(url, headers=request_header).json()
```
